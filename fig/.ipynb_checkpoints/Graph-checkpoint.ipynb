# 实验4

```python
def print_graph(x, y_list, model_list, xlabel="Epochs", ylabel="", title=""):
    assert len(y_list) == len(model_list)
    lines = []
    for k, model in enumerate(model_list):
        y = y_list[k]
        p, = plt.plot(x, y)
        lines.append(p )
    plt.legend(lines, model_list)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
```

## Train Loss

```python
models = ["exp4_overfit_student_alpha{}".format(i) for i in [2,5,8,10]  ]
y_list = []
for k, model in enumerate(models):
    data_csv = pd.read_csv( "csvlogs/{}.csv".format(model) )
    x = data_csv['epoch'].as_matrix()
    y_list.append( data_csv['loss'].as_matrix() )
print_graph(x, y_list, models, "Epochs", "Train Loss", "Train Loss")
```

## Val ACC

```python
models = ["exp4_overfit_student", "exp4_normalfit_student"]
plt.figure(figsize = (15,5))
for k, model in enumerate(models):
    alpha_model_names = ["{}_alpha{}".format(model, alpha) for alpha in [0.2, 0.5, 0.8]]
    y_list = []
    for alpha in [2, 5, 8]:
        data_csv = pd.read_csv( "csvlogs/{}_alpha{}.csv".format(model, alpha) )
        x = data_csv['epoch'].as_matrix()
        y_list.append( data_csv['val_categorical_accuracy'].as_matrix() )
    plt.subplot(120+k+1)
    print_graph(x, y_list, alpha_model_names, "Epochs", "Val ACC", "{} Val ACC".format(model))
```

## Val Loss

graph on right removes the outlier before epoch 25

```python
models = ["exp4_overfit_student", "exp4_normalfit_student"]
plt.figure(figsize = (15,5))
for k, model in enumerate(models):
    alpha_model_names = ["{}_alpha{}".format(model, alpha) for alpha in [0.2, 0.5, 0.8]]
    y_list = []
    for alpha in [2, 5, 8]:
        data_csv = pd.read_csv( "csvlogs/{}_alpha{}.csv".format(model, alpha) )
        x = data_csv['epoch'].as_matrix()
        y_list.append( data_csv['val_loss'].as_matrix() )
    plt.subplot(120+k+1)
    print_graph(x, y_list, alpha_model_names, "Epochs", "Val Loss", "{} Val Loss".format(model))
```

## 选出各模型最优的alpha by val loss

```python
for model in ["exp4_overfit_student", "exp4_normalfit_student"]:
    min_loss_by_model = 9999
    max_acc_by_model = 0
    best_alpha_by_loss = 0
    best_alpha_by_acc = 0
    
    for alpha in [2, 5, 8]:
        data_csv = pd.read_csv( "csvlogs/{}_alpha{}.csv".format(model, alpha) )
        if np.min(data_csv['val_loss'].as_matrix()) < min_loss_by_model:
            min_loss_by_model = np.min(data_csv['val_loss'].as_matrix())
            best_alpha_by_loss = alpha
        if np.max(data_csv['val_categorical_accuracy'].as_matrix()) > max_acc_by_model:
            max_acc_by_model = np.max(data_csv['val_categorical_accuracy'].as_matrix())
            best_alpha_by_acc = alpha

    print("{} - {} - alpha={} - {} - alpha={}".format(model, min_loss_by_model, best_alpha_by_loss/10, max_acc_by_model, best_alpha_by_acc/10))
```

### 模型对比

```python
models = ["exp4_overfit_student", "exp4_normalfit_student"]
y_list = []
plt.figure(figsize=(20,5))
for model in models:
    if model == "exp4_overfit_student":
        alpha = 2
    else:
        alpha = 8    
    data_csv = pd.read_csv( "csvlogs/{}_alpha{}.csv".format(model, alpha) )
    x = data_csv['epoch'].as_matrix()
    y_list.append( data_csv['val_loss'].as_matrix() )

plt.subplot(121)
print_graph(x, y_list, models, "Epochs", "Val Loss", "Val Loss")

plt.subplot(122)
print_graph(x[50:], [y[50:] for y in y_list], models, "Epochs", "Val Loss", "Val Loss")
```
